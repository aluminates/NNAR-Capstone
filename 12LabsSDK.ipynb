```python
excel_file = '/content/drive/MyDrive/12Labs/video_details.xlsx'
video_folder="/content/drive/MyDrive/Datasets - Twelve labs/Violence & Non violence (+ weaponised)/SCVD/SCVD_converted/Train/Violence/*.avi"
index_id='66c611d3caeae2951c46d160'

video_files = glob(video_folder) # Example: "/videos/*.mp4
file_names = [
    "v071_converted", "v075_converted", "v076_converted", "v056_converted", "v20_converted",
    "v001_converted", "v002_converted", "v006_converted", "v007_converted", "v008_converted",
    "v009_converted", "v011_converted", "v012_converted", "v051_converted", "v052_converted",
    "v057_converted", "v_061_converted", "v_063_converted", "v_064_converted", "v_067_converted",
    "v_073_converted", "v_076_converted", "v_077_converted", "v_078_converted", "v_080_converted",
    "v_082_converted", "v_085_converted", "v_086_converted", "v_087_converted", "v_090_converted",
    "w_002_converted"
]
corrected_file_names = [
    "v020_converted", "v061_converted", "v063_converted", "v064_converted", "v067_converted",
    "v073_converted", "v076_converted", "v077_converted", "v078_converted", "v080_converted",
    "v082_converted", "v085_converted", "v086_converted", "v087_converted", "v090_converted"
]

# Get all video files in the folder
video_files = glob(video_folder)
print
# Filter the video files to include only the specific names
selected_videos = [video for video in video_files if os.path.splitext(os.path.basename(video))[0] in corrected_file_names]
print(selected_videos)
# Now 'selected_videos' contains only the files with the specific names
for video_file in selected_videos:
  print(f"Uploading {video_file}")
  task = client.task.create(index_id=index_id, file=video_file, language="en")
  print(f"Task id={task.id}")

  # (Optional) Monitor the video indexing process
  # Utility function to print the status of a video indexing task
  def on_task_update(task: Task):
          print(f"  Status={task.status}")
  task.wait_for_done(callback=on_task_update)
  if task.status != "ready":
      raise RuntimeError(f"Indexing failed with status {task.status}")
  if os.path.exists(excel_file):
    # Load the existing Excel file
    df_existing = pd.read_excel(excel_file)
    # Create a DataFrame with the new data
    df_new = pd.DataFrame({'Video ID': [task.video_id], 'File Path': [video_file]})
    # Concatenate the existing data with the new data
    df = pd.concat([df_existing, df_new], ignore_index=True)
  else:
    # Create a new DataFrame
    df = pd.DataFrame({'Video ID': [task.video_id], 'File Path': [video_file]})
  df.to_excel(excel_file, index=False)
  print(f"Uploaded {video_file}. The unique identifer of your video is {task.video_id}.")
```

## Below code is to get a text output from the model given a video id and a prompt

```python
promptt='''Carefully watch the provided video and generate a caption by following these steps:

Initial Observation: First, watch the entire video to get a sense of the overall scene and the sequence of events. Identify the main actions and interactions.

Break Down Key Actions: Segment the video into smaller parts. For each part, describe the specific actions taking place, such as movements, interactions between individuals, or any noticeable changes in the environment.

Detail the Sequence: Capture the order in which actions occur. Describe each step clearly, ensuring that the sequence of events is logical and accurate.

Focus on Observable Details: Stick to what you can directly observe in the video. Avoid assumptions or interpretations about emotions or motives. Describe actions factually, such as ‘a person raises their arm’ rather than ‘a person seems angry.’

Contextual Clarity: Mention any relevant context visible in the video, such as the setting or objects involved, but only if it directly contributes to understanding the scene.

Summarize Accurately: Summarize the actions and interactions in a concise caption that accurately reflects the events in the video, contributing to a solid semantic understanding.

Review and Refine: Finally, review the caption to ensure it’s logical, clear, and accurately represents the observed events without unnecessary details.'''
```
